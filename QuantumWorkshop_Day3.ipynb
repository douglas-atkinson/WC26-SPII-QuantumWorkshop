## Quantum State and Circuits

### Sampling from a Probability Distribution

- Each run of a quantum computation produces a single probabilistic outcome.
- Repeated measurements produce samples fro a probability distribution determined by the amplitudes of the quantum state.
- By manipulating amplitudes with quantum operations, algorithms can shape the distribution to sample useful outcomes, such as uniform random values.

### Quantum computations with Python Simulators
- Classical simulators model quantum states and operations, allowing us to develop, test, and inspect quantum programs before running them on real hardware.
- Simulation is computationally expensive because the required processing doubles with each additional qubit.


## Python Quantum SDK Overview
Tools and libraries are available in the form of SDKs (software development kits) for designing, simulating, and running quantum circuits using Python.
- These frameworks allow users to define qubits, apply quantum gates, execute circuits on simulators or real quantum hardware, and analyze measurement results.
- Examples include:
  - Qiskit (https://quantum.cloud.ibm.com/docs/en/guides/install-qiskit): an open-source SDK developed by IBM for building and running quantum circuits on simulators or IBM quantum hardware.
  - Cirq (https://quantumai.google/cirq): a Python framework from Google designed for constructing and executing quantum circuits, particularly for NISQ-era devices.
  - Amazon Braket (https://amazon-braket-sdk-python.readthedocs.io/en/stable/): a Python library that allows users to design circuits and run them on simulators or quantum hardware from multiple providers through AWS.
  - PennyLane (https://pennylane.ai/install): a framework for hybrid quantum-classical computation, often used in quantum machine learning research.
- The SDKs let developers define qubits, apply gates, run circuits on simulators or real hardware, and analyze results while abstracting the underlying hardware complexity.


## States, Gates, and Oracles

In quantum computing, **states** describe the current condition of a qubit or group of qubits, typically represented as vectors in a mathematical space.

These states encode the amplitudes—and therefore the probabilities—of possible measurement outcomes.

**Gates** are the operations that transform these states, similar to how logic gates operate on classical bits.

Each quantum gate applies a specific linear transformation, changing the amplitudes and phases of the state without directly revealing its value.

**Oracles** are specialized quantum subroutines—built from gates—that encode a problem or condition into a circuit.

Instead of revealing answers directly, an oracle “marks” certain states (often by flipping a phase or writing to an auxiliary qubit), allowing later steps in the algorithm to amplify or extract the desired outcomes.

Together, states, gates, and oracles form the foundation of how quantum circuits represent problems, manipulate information, and guide computations toward useful results.

## Parameterized Circuits and Reuse

Parameterized circuits are quantum circuits that include adjustable values (parameters) within certain gates, allowing the same circuit structure to be reused with different numerical settings.
- Instead of building a new circuit each time, we define a flexible template and vary the parameters to explore different behaviors or optimize performance. - This approach is especially useful in hybrid quantum-classical algorithms, where classical optimization routines update parameters iteratively to improve results.
- Parameterization promotes reuse, modular design, and efficient experimentation without redesigning the entire circuit.

## Hybrid Quantum–Classical Workflows
Hybrid quantum–classical workflows combine quantum circuits with classical computation to solve problems more effectively than either approach alone.
- In these workflows, a quantum processor performs tasks suited to quantum mechanics—such as preparing and transforming complex superposition states—while a classical computer handles optimization, parameter updates, and control logic.
- The two systems communicate iteratively, with classical algorithms analyzing measurement results and adjusting quantum circuit parameters in a feedback loop.
- This collaboration is central to many near-term quantum algorithms, especially on today’s noisy, intermediate-scale quantum (NISQ) devices.
